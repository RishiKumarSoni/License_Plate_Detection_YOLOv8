```python
import os
from PIL import Image
```

```python
def change_image_format(directory, new_format):
    """
    Change the file extension of all images in a directory to a specified format.
    
    Parameters:
    - directory: The path to the directory containing the images.
    - new_format: The desired new file format (e.g., 'jpg', 'png').
    """
    count = 0
    for filename in os.listdir(directory):
        if filename.endswith(('.jpeg', '.jpg', '.png', '.gif', '.bmp')):
            # Open the image file
            image_path = os.path.join(directory, filename)
            try:
                img = Image.open(image_path)
            except Exception as e:
                print(f"Failed to open {filename}: {e}")
                continue
            
            # Generate new filename with the specified format
            new_filename = "image" + str(count) + '.' + new_format
            
            # Save the image with the new filename and format
            new_image_path = os.path.join(r"C:\Users\sonir\Documents\rishi\cdac_mohali\sem2\CV\ANPR\renamed", new_filename)
            try:
                img.save(new_image_path)
                print(f"Converted {filename} to {new_format} format.")
            except Exception as e:
                print(f"Failed to save {new_filename}: {e}")
    
        count = count + 1
```

```python
def Rename(directory):
    count = 0

    for file in os.listdir(directory):

        image_path = os.path.join(directory, file)
        img = Image.open(image_path)

        new_filename = "image" + str(count) + '.' + "jpg"

        new_image_path = os.path.join(r"C:\Users\sonir\Documents\rishi\cdac_mohali\sem2\CV\ANPR\renamed", new_filename)

        try:
            img.save(new_image_path)
            print("file renamed")
        except Exception as e:
            print(f"error:-  {e}")
    
    count += 1
```

```python
# Example usage
if __name__ == "__main__":
    # Specify the directory containing the images and the desired new format
    directory_path = r"C:\Users\sonir\Documents\rishi\cdac_mohali\sem2\CV\ANPR\images"
    new_format = 'jpg'
    
    # Change the image format
    change_image_format(directory_path, new_format)
```
